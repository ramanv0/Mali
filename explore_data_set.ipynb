```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python

filename = 'dataset/bodmas.npz'
data = np.load(filename)
X = data['X']  # all the feature vectors
y = data['y']  # labels, 0 as benign, 1 as malicious
```

```python
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
train_df = pd.DataFrame(X_train, columns=np.arange(X_train.shape[1]))
train_df['label'] = y_train
```

```python
train_df.head()
```

```python
predictor = TabularPredictor(label = 'label', path = 'models/ag_test')
predictor.fit(train_df)
```

```python
dev_df = pd.DataFrame(X_test, columns=np.arange(X_test.shape[1]))
dev_df['label'] = y_test
```

```python
predictor.leaderboard(dev_df)
```
